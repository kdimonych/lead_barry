### Initial Conditions

```python
from IPython.display import display, Markdown

MainVOut = 9  # V, the main output voltage
SecondaryVOut = 12  # V, the secondary output voltage (switchable between 9V and 12V)
Vref = 1.25  # V, the reference voltage of the XL6019 regulator
R1 = 325  # Ohms, the resistance of R1

display(
    Markdown(
        f"The main output voltage is $V_{{out, main}} = {MainVOut}\\,\\text{{V}}$ and the secondary output voltage is $V_{{out, secondary}} = {SecondaryVOut}\\,\\text{{V}}$."
    )
)
display(
    Markdown(
        f"The reference voltage of the XL6019 regulator is $V_{{ref}} = {Vref}\\,\\text{{V}}$."
    )
)
display(Markdown(f"The resistance of R1 is $R_1 = {R1}\\,\\text{{Ohms}}$."))
```

> The R19 for 9V is equal: 
> \begin{equation}
> R_{19} = R_1 \cdot (\frac{V_{out\,main}}{V_{ref}} - 1)
> \end{equation}

```python
R19 = R1 * ((MainVOut / Vref) - 1)

display(Markdown(f"The resistance of R19 is $R_{{19}} = {R19:.2f}\\,\\text{{Ohms}}$."))
```

> The equivalent resistance od $R_{eq}$ 0f two parallel connected $R_1$ and $R_{20}$ can be derived from:
>
> \begin{equation}
> V_{out, secondary} = V_{ref} \cdot (1 + \frac{R_{19}}{R_{eq}}) \implies R_{eq} = \frac{R_{19} \cdot V_{ref}}{V_{out, secondary}-V_{ref}}
> \end{equation}
>
> \begin{equation}
> R_{20} = \frac{R_{eq} \cdot R_1}{R_1-R_{eq}}
> \end{equation}

```python
Req = R19 * Vref / (SecondaryVOut - Vref)

display(
    Markdown(
        f"The equivalent resistance of R19 and the internal resistor of the XL6019 regulator is $R_{{eq}} = {Req:.2f}\\,\\text{{Ohms}}$."
    )
)

R20 = Req * R1 / (R1 - Req)

display(Markdown(f"The resistance of R20 is $R_{{20}} = {R20:.2f}\\,\\text{{Ohms}}$."))
```

## Result

```python
display(Markdown(f"The resistance of R19 is $R_{{19}} = {R19:.2f}\\,\\text{{Ohms}}$."))

display(Markdown(f"The resistance of R20 is $R_{{20}} = {R20:.2f}\\,\\text{{Ohms}}$."))
```
