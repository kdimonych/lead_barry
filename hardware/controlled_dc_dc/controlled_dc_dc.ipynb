### Initial Conditions

```python
from IPython.display import display, Markdown

MinVout = 5  # V, the main output voltage
MaxVout = 15  # V, the secondary output voltage (switchable between 9V and 12V)
Vref = 1.25  # V, the reference voltage of the XL6019 regulator
R1 = 1000  # Ohms, the resistance of R1
Rds = 6  # Ohms, the on-resistance of the MOSFET
Fs = 1000000  # Hz, the switching frequency of the regulator
n = 0.3  # Filter factor (30% of the switching frequency)

display(
    Markdown(
        f"The minimum output voltage is $V_{{out, min}} = {MinVout}\\,\\text{{V}}$ and the maximum output voltage is $V_{{out, max}} = {MaxVout}\\,\\text{{V}}$."
    )
)
display(
    Markdown(
        f"The reference voltage of the XL6019 regulator is $V_{{ref}} = {Vref}\\,\\text{{V}}$."
    )
)
display(Markdown(f"The resistance of R1 is $R_1 = {R1}\\,\\text{{Ohms}}$."))
display(
    Markdown(f"The on-resistance of the MOSFET is $R_{{ds}} = {Rds}\\,\\text{{Ohms}}$.")
)
display(
    Markdown(
        f"The switching frequency of the regulator is $F_s = {Fs / 1000}\\,\\text{{kHz}}$ and the filter factor is $n = {n}$."
    )
)
```

> The R19 for 9V is equal: 
> \begin{equation}
> R_{19} = R_1 \cdot (\frac{V_{out\,min}}{V_{ref}} - 1)
> \end{equation}

```python
R19 = R1 * ((MinVout / Vref) - 1)

display(Markdown(f"The resistance of R19 is $R_{{19}} = {R19:.2f}\\,\\text{{Ohms}}$."))
```

> The equivalent resistance od $R_{eq}$ of two parallel connected $R_1$ and $R_{20}$ can be derived from:
>
> \begin{equation}
> V_{out, max} = V_{ref} \cdot (1 + \frac{R_{19}}{R_{eq}}) \implies R_{eq} = \frac{R_{19} \cdot V_{ref}}{V_{out, max}-V_{ref}}
> \end{equation}
>
> \begin{equation}
> R_{20} = \frac{R_{eq} \cdot R_1}{R_1-R_{eq}} - R_{ds}
> \end{equation}

```python
Req = R19 * Vref / (MaxVout - Vref)

display(
    Markdown(
        f"The equivalent resistance of R19 and the internal resistor of the XL6019 regulator is $R_{{eq}} = {Req:.2f}\\,\\text{{Ohms}}$."
    )
)

R20 = Req * R1 / (R1 - Req) - Rds

display(Markdown(f"The resistance of R20 is $R_{{20}} = {R20:.2f}\\,\\text{{Ohms}}$."))
```

### Filter

> The time constant of RC-filter has to satisfy the condition $T \ll \tau$ where
>
> \begin{equation}
> \tau = Req \cdot C_{13}
> \end{equation}
>
> Let use some $\tau$ to $T$ factor $0 < n < 1$, so:
>
> \begin{equation}
> T = \tau \cdot n \implies \frac {T}{n} = \frac {1}{F_s \cdot n} 
> \end{equation}
>
> The filter capacitor:
>
> \begin{equation}
> C_{13} = \frac {\tau}{R_{eq}} = \frac {1}{R_{eq} \cdot F_s \cdot n}
> \end{equation}

<div style="text-align: center;">
<img src="./rc_filter.drawio.svg" alt="Equivalent RC-filter sircuit" style="display: block; margin: 0 auto; width: 30%;"/>
</div>

> Where:
>
> \begin{equation}
> R_{eq} = R_{1} || R_{19} || (R_{20} + R_{ds})
> \end{equation}

```python
Req = (R1 ** (-1) + R20 ** (-1) + (Rds + R20) ** (-1)) ** (-1)

display(
    Markdown(
        f"The equivalent resistance of R1, R20, and the on-resistance of the MOSFET is $R_{{eq}} = {Req:.2f}\\,\\text{{Ohms}}$."
    )
)

C13 = 1 / (Req * Fs * n)

display(
    Markdown(f"The capacitance of C13 is $C_{{13}} = {C13 * 1e9:.0f}\\,n \\text{{F}}$.")
)
```

## Result

```python
display(Markdown(f"The resistance of R19 is $R_{{19}} = {R19:.2f}\\,\\text{{Ohms}}$."))

display(Markdown(f"The resistance of R20 is $R_{{20}} = {R20:.2f}\\,\\text{{Ohms}}$."))

display(
    Markdown(f"The capacitance of C13 is $C_{{13}} = {C13 * 1e9:.0f}\\,n \\text{{F}}$.")
)
```
