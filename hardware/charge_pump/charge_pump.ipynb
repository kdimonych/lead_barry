## Problem Statement
> In order to correctly drive the top-side N-type MOSFET switch, the circuit required to have voltage source $V_{Drive}$ up to $V_{GS}-V_{DS(On)}$ volts above the main source voltage $V_{CC}$, where $V_{GS}$ - Gate-to-Source Threshold Voltge, $V_{DS(On)}$ - Drain to Source voltage when transistor is open.
>
>
> \begin{equation}
> V_{Drive} = V_{CC} + V_{GS}-V_{DS(On)}
> \end{equation}

### Estimate the maximum possible voltage increase

To estimate the maximum theoretical increase of voltage, let's review the idealized switch equivalent schema:

<div style="text-align: center;">
<img src="./charge_pump_equivalent_schema.drawio.svg" alt="N-Type MOSFET Usage" style="display: block; margin: 0 auto; width: 50%;"/>
</div>

At the first stage the capacitor C1 is being charged up to $V_{CC}$ voltage. At the second stage the C1 is being discharged to C2. 

At the second stage both capacitors form voltage divider C1,C2:

<div style="text-align: center;">
<img src="./cap_voltage_devider.drawio.svg" alt="N-Type MOSFET Usage" style="display: block; margin: 0 auto; width: 50%;"/>
</div>

, where $V_{eq}$ - is equivalent voltage which is equal to:

> \begin{equation}
> V_{eq} = V_{CC} + V_{C1} - V_{C2}\text{.}
> \end{equation}

So, the new voltage at C2 $V_{C2(new)} = V_{eq} \dot G$, where $G$ is reduction gain of the divider:

> \begin{equation}
> G = \frac{Z_{C2}}{Z_{C1} + Z_{C2}} = \frac{\frac 1{\omega j C2}}{\frac 1{\omega j C1} + \frac 1{\omega j C2}} = \frac{\frac 1{\omega j C2}}{\frac{\omega j C2+\omega j C2}{(\omega j)^2 C1 \cdot C2}}=\frac{C2}{C2+C1}
> \end{equation}
 
Substitute:

> \begin{equation}
> V_{C2(new)} = V_{eq} \frac{C2}{C2+C1}
> \end{equation}
