### Estimate the maximum possible voltage increase

At the stage above the both capacitors form voltage divider C1, C2 connected to the voltage source $V_{CC}$:

<div style="text-align: center;">
<img src="./cap_voltage_divider.drawio.svg" alt="N-Type MOSFET Usage" style="display: block; margin: 0 auto; width: 20%;"/>
</div>

The  Kirchhoff’s Voltage Law (KVL), which states that the algebraic sum of all electrical potential differences (voltages) around any closed loop in a circuit is equal to zero. So, the equation of a circuit above is:

$$
V_{CC} + V_{C1} + V_{C2} = 0
$$

This equation is true for a settled equilibrium, after the transient is done. But if we take into account the momentary charge polarity of all the sources at the initial moment

<div style="text-align: center;">
<img src="./cap_voltage_divider_polarity.drawio.svg" alt="N-Type MOSFET Usage" style="display: block; margin: 0 auto; width: 20%;"/>
</div>

, and the constraint that $V_{C2}(0) < V_{CC} + V_{C1}(0)$ since it is the initial moment of stage 2 and charges are not equlized yet, we get the following contradiction: 

$$
V_{CC} + V_{C1}(0) - V_{C2}(0) = 0 \implies V_{CC} + V_{C1}(0) = V_{C2}(0)
$$

hence, to equalize the equation, there must be one more term, which is the counterpart to the sum $V_{CC}(0) + V_{C1}(0) - V_{C2}(0)$, let it be the $\Delta V(0)$

$$
V_{CC}(0) + V_{C1}(0) - V_{C2}(0) - \Delta V(0) = 0 \implies \Delta V(0) = V_{CC}(0) + V_{C1}(0) - V_{C2}(0)
$$

To calculate the voltage rise at C2 at the end of the transient process, we can treat the circuit as if it is simple capacitor divider with a voltage source $\Delta V(0)$ applied to it:

<div style="text-align: center;">
<img src="./cap_voltage_divider_delta_v.drawio.svg" alt="N-Type MOSFET Usage" style="display: block; margin: 0 auto; width: 20%;"/>
</div>

The voltage drop at C2 at the end of the transition is:

$$
\Delta V_{C2}(5\tau) \approx \Delta V(0) \cdot G
$$

, where $\tau$ - is a time constant of transient, $G$ is divider gain:

$$
G = \frac{Z_{C2}}{Z_{C1} + Z_{C2}} = \frac{\frac 1{\omega j C2}}{\frac 1{\omega j C1} + \frac 1{\omega j C2}} = \frac{\frac 1{\omega j C2}}{\frac{\omega j C1 + \omega j C2}{(\omega j)^2 C1 \cdot C2}}=\frac{\frac 1{\omega j C2} \cdot (\omega j)^2C1 \cdot C2}{\omega j C2 + \omega j C1}=\frac{C1}{C2+C1}
$$
 
Substitute:

$$
\Delta V_{C2}(5\tau) \approx \Delta V(0) \frac{C1}{C2+C1} = \left(V_{CC} + V_{C1}(0) - V_{C2}(0) \right) \cdot \frac{C1}{C2+C1}
$$

and full voltage at C2: 

$$
V_{C2}(5\tau) = \Delta V_{C2}(5\tau) + V_{C2}(0)
$$

Hence, according to Kirchhoff’s Voltage Law (KVL):

$$
V_{C1}(5\tau) = -V_{CC} - V_{C2}(5\tau) = -V_{CC} - V_{C2}(0) - \Delta V_{C2}(5\tau) = V_{CC} - V_{C2}(0) - \left(V_{CC} + V_{C1}(0) - V_{C2}(0)\right) \cdot \frac{C1}{C2+C1}
$$

Prove that voltages will be eualized:

$$
V_{CC} + V_{C1}(5\tau) + V_{C2}(5\tau) = 0
$$

, substitute:

$$
V_{CC} - V_{CC} - V_{C2}(0) - \Delta V_{C2}(5\tau) + \Delta V_{C2}(5\tau) + V_{C2}(0) = 0
$$
